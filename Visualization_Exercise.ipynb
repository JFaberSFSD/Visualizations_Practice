```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
import seaborn as sns

df = pd.read_excel('/content/drive/MyDrive/04 - Week 4 (July 11th)/Favorite Sports Data.xlsx')

df.head()
```

```python
df_new = df.melt(id_vars='Sport')

df_new = df_new.rename(columns={'variable':'Year', 'value':'Total_Percent'})

df_new
                      
```

```python
fig, ax = plt.subplots(figsize=(16,8))
ax = sns.lineplot(data=df_new, x='Year', y='Total_Percent', hue='Sport',marker='*',linewidth=4)

ax.set_title("American's Favorite Sport Since 2004",fontsize=28,fontweight=4)

ax.set_xlabel(ax.xaxis.get_label().get_text(),
              fontsize='x-large')
ax.set_ylabel(ax.yaxis.get_label().get_text(),
              fontsize='x-large')

ax.set_ylabel('Total Percent',fontsize=22)
ax.set_xlabel('')


def percentage (x,pos):
    
    return f"{x:,.2f}%"


from matplotlib.ticker import FuncFormatter
percentage = FuncFormatter(percentage)

ax.yaxis.set_major_formatter(percentage)

sns.set_context('paper')

plt.xticks(fontsize=18)
plt.yticks(fontsize=14)

plt.style.use('dark_background')
plt.setp(ax.get_legend().get_title(), fontsize='20') 
plt.setp(ax.get_legend().get_texts(), fontsize='16') 
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
plt.show()
```
